```python
encoded_track = one_hot(df, 'Track')
encoded_track
```

```python
encoded_country = one_hot(df, 'Country')
encoded_country.shape
```

```python
encoded_driver = one_hot(df, "Driver")
encoded_driver.shape
```

```python
encoded_team = one_hot(df, 'Team')
encoded_team.shape
```

```python
numerical_df = df[['Year', 'Position', 'Starting Grid', 'Points', 'Pitstop Time', 'Weather', 'avg_confidence_score', 'Winner']]

categorical_columns = [encoded_track, encoded_country, encoded_driver, encoded_team]

final_df = pd.concat([numerical_df] + categorical_columns, axis=1)


final_df
```
