```python
'''def one_hot(x):
  threshold = 100
  counts = x.value_counts()

  repl = counts[counts <= threshold].index

  ohe = pd.get_dummies(x.replace(repl, 'uncommon')).astype(int)
  return ohe'''

def one_hot(x):
  threshold = 100
  counts = x.value_counts()

  repl = counts[counts <= threshold].index
  x = x.replace(repl, 'uncommon')
  x = x.values.reshape(-1,1)

  encoder = OneHotEncoder(drop = 'first', sparse_output = False)
  ohe = encoder.fit_transform(x)
  return ohe
```

```python
encoded_track = one_hot(df['Track'])
encoded_track.shape
```

```python
encoded_country = one_hot(df['Country'])
encoded_country.shape
```

```python
encoded_driver = one_hot(df['Driver'])
encoded_driver.shape
```

```python
encoded_team = one_hot(df['Team'])
encoded_team.shape
```

```python
final_making = np.hstack((df[['Year']].values, encoded_track))
final_making = np.hstack((final_making, encoded_country))
final_making = np.hstack((final_making, df[['Position']]))
final_making = np.hstack((final_making, encoded_driver))
final_making = np.hstack((final_making, encoded_team))
final_making = np.hstack((final_making, df[['Starting Grid']]))
final_making = np.hstack((final_making, df[['Points']]))
final_making = np.hstack((final_making, df[['Pitstop Time']]))
final_making = np.hstack((final_making, df[['Weather']]))
final_making = np.hstack((final_making, df[['avg_confidence_score']]))
final_making = np.hstack((final_making, df[['Winner']]))

final_making

final_df = pd.DataFrame(final_making)
final_df
```
